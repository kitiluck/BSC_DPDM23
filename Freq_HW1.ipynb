## 1. ของช่วงเวลา Christmas & New year (1 dec - 30 dec)

```python
#1-31 Dec
xmas_new = data[(data['Date'].dt.month == 12) & (data['Date'].dt.day >= 1) & (data['Date'].dt.day <= 31)]
print(xmas_new)
```

```python
xmas_new
```

```python
trans_xmas_new = xmas_new.groupby('InvoiceNo')['Description'].apply(list).reset_index()
print(trans_xmas_new)
```

```python
trans_newxmas_apply = trans_xmas_new['Description'].apply(lambda x: [str(item) for item in x])
trans_newxmas_apply
```

```python
min_sup = 0.036 #เปลี่ยนจนกว่าจะเจอ...

frequentItemsets_newxmas = list(apriori(trans_newxmas_apply, min_support = min_sup))
for i in frequentItemsets_newxmas:
    if i[0].__len__()>1:
        print('yes')
        print(i)
```
