## 1. ของช่วงเวลา Christmas & New year (1 dec - 30 dec)

```python
#Select 1-31 Dec
Dec= data[(data['Date'].dt.month == 12) & (data['Date'].dt.day >= 1) & (data['Date'].dt.day <= 31)]

#printcheck
print(Dec)
```

```python
Dec
```

```python
transac1 = Dec.groupby('InvoiceNo')['Description'].apply(list).reset_index()
print(transac1)
```

```python
transac1apply = transac1['Description'].apply(lambda x: [str(item) for item in x])
transac1apply
```

```python
min_sup = 0.036 #เปลี่ยนจนกว่าจะเจอ...

frequentItemsets = list(apriori(transac1apply,min_support = min_sup))
for i in frequentItemsets:
    if i[0].__len__()>1:
        print('yes')
        print(i)
```

```python
frequentItemset = list(apriori(transac1apply, min_support=0.036))

for record in frequentItemset:
    if len(record.items) > 1:
        print('yes')
        print(f"Items: {', '.join(record.items)}")
        print(f"Support: {record.support:.4f} ({record.support*100:.2f}%)")

        for i, stat in enumerate(record.ordered_statistics, 1):
            items_base = 'None' if not stat.items_base else ', '.join(stat.items_base)
            items_add = ', '.join(stat.items_add)
            confidence = stat.confidence
            lift = stat.lift

            print(f"Rule {i}: ({items_base} -> {items_add} : Confidence={confidence:.4f}, Lift={lift:.4f})")
        print()
```
